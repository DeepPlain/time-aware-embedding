```python
import numpy as np
```

```python
def load_pretrained_glove(file):
    print("Loading")
    model = {}
    f = open(file, 'r')
    for line in f:
        splitLine = line.split()
        word = splitLine[0]
        embedding = np.array([float(val) for val in splitLine[1:]])
        model[word] = embedding
    print(len(model),"words loaded")
    return model
```

```python
model = load_pretrained_glove("dataset/glove.6B.300d.txt")
```
